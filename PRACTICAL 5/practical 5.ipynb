Q3. Two-tailed Z-test (Comparison Problem)
Given:
Population mean (μ₀) = 12
Population SD (σ) = 5.5
Sample mean (x̄) = 10.7
Sample size (n) = 36

Step 1 — Hypotheses

Null Hypothesis H₀:
μ = 12
(The regional players' average points are not different from the historic mean)

Alternative Hypothesis H₁:
μ ≠ 12
(The regional players' average points are different from the historic mean)

```python
# Step 2 — Compute Z-score
import numpy as np
mu0 = 12
sigma = 5.5
xbar = 10.7
n = 36

z = (xbar - mu0) / (sigma / np.sqrt(n))
z
```

```python
# Step 3 — Two-tailed p-value
import scipy.stats as st

p_value = 2 * (1 - st.norm.cdf(abs(z)))
p_value
```

Interpretation
If p < 0.05, reject H₀ → scores are significantly different
If p ≥ 0.05, fail to reject H₀ → no significant difference
